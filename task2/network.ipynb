```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/arstotrix/semantics_and_whatnot/main/task2/shifts.csv")
```

```python
G=nx.from_pandas_edgelist(df[['word_1', 'word_2', 'strength', 'shift_type']], 'word_1', 'word_2', edge_attr=True)
```

```python
pos = nx.spring_layout(G, seed=11) 
```

```python
nx.draw(G, pos, with_labels=True, font_weight='bold')
plt.show()
```
